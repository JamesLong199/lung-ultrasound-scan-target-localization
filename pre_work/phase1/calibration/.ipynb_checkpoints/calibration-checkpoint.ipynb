## Setup
A paper containing checkerboard pattern is attached to a fixed flat surface

25 images containing the checkerboard pattern are taken, capturing the radial and tangential distortions of the camera.

## Loading and Displaying Image

4 images were selected to be displayed with inner corners annotated. 

```python
from CamCalib import CamCalib
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
myCam = CamCalib(w=8,h=6)
# load images
img_dir = 'images'

fig,axes = plt.subplots(2,2,figsize=(30,15))

for i,img_name in enumerate(os.listdir(img_dir)):
    fname = img_dir + '/' + img_name
    marked_img = myCam.add_image(fname,plt=True)
    
    # display 4 annotated images
    if i%6 == 0:
        axes.ravel()[int(i/6)-1].imshow(marked_img)

        
```

## Calibrate Camera

'retval': root mean square reprojection error in terms of pixels, between 0.1 and 1.0 for a good calibration

'cameraMatrix': intrinsic matrix of the camera

'distCoeffs': distortion coefficients of the camera

```python
# calibrate camera
retval, cameraMatrix, distCoeffs, rvecs, tvecs = \
myCam.run_calibration()

np.set_printoptions(precision=3)
# root mean square reprojection error in terms of pixels
# between 0.1 and 1.0 for a good calibration
print('retval:',retval)
print('cameraMatrix:\n',cameraMatrix)
print('distCoeffs:\n',distCoeffs)
```

## Undistort Camera Images

With distortion coefficients obtained, we can undistort the images displayed above

```python
img_dir = 'images'

fig,axes = plt.subplots(2,2,figsize=(30,15))

for i,img_name in enumerate(os.listdir(img_dir)):
    fname = img_dir + '/' + img_name
    restored_img = myCam.run_undistortion(fname,plt=True)
    
    # display 4 annotated images
    if i%6 == 0:
        axes.ravel()[int(i/6)-1].imshow(restored_img)
```
